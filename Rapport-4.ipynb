### Sammendrag:
i denne rapporten fokuserer vi på et forsøke hvor en kompakt kule slippes gjennom numerisk og eksperimentell bane. Først er det en teoriutledning, etterfulgt av metode. Den numeriske delen av forsøket blir kodet, mens den eksperimentelle delen filmet og beregnes i bakgrunn av programmet tracker. Resultatene for forsøket endte opp med å bli middelverdi og standardfeil for sluttfart $1,108 \pm 0,0219$ , total rulletid  $2,410 \pm 0,0958$ , total kinetisk energi  $0,0191 \pm 0,000754$  og tap av mekanisk energi  $0,0132 \pm 0,000754$. Til slutt følger en diskusjonsdel av mulige feilkilder og en konklusjon.

### Introduksjon:
Motivasjonen bak forsøket grunner i Newtons 2. lov $\sum F = {ma}$ (Grøn,2021) og bevaring av mekanisk energi. I dette forsøket har vi sett på en kule som ruller rent langs en ferdigsimulert bane fra en oppgitt kode. Akselerasjon og krefter vil endre seg grunnet krummet underlag. Hensikten med eksperimentet er å måle de nummeriske verdiene, sluttfart, total rulletid, tap av mekanisk energi og total kinetisk energi ved slutten og deres representative middelverdi og standardfeil. Videre sammenliknes de numerisk og eksperimentelle verdiene. Deretter gjennomfører vi en usikkerhetsanalyse for å drøfte og se forskjellen mellom eksperimentet og nummerikken.


### Teori:

#### Energibevaring og mekanisk energi:
Mekanisk energi i et lukket system med ren rulling vil være bevart. For å finne den totale mekaniske energien, må vi ta utgangspunkt i kinetisk og potensiell energi (NTNU,u.å). Kinetisk energi med hensyn på rulling vil være gitt som:


\begin{equation}
\tag{1}
K = {{1+c\over 2}mv^2}.
\end{equation}

Summen av rotasjonsenergien $cmv^2\over 2$ og translasjonsenergien $mv^2\over 2$ tilsvarer total kinetisk energi for systemet. $m$ er massen til objektet og $c$ får vi fra objektets treghetsmoment(Støvneng,2023,s.1). 

Den totale mekanisk energien vil da bli:

\begin{equation}
\tag{2}
E = mgy(x)-{{1+c\over 2}mv^2}
\end{equation}

hvor $y(x)$ er kulens posisjon langs x-aksen og $v$ er hastigheten. $g$ er tyngdeakselerasjonen gitt som $9.81m/s^2$.

Tar vi for oss energibevarelse, vil hastigheten være oppgitt som:

\begin{equation}
\tag{3}
v=\sqrt{10\cdot g(y_0-y)\over 7}.
\end{equation}

Her er $c={2\over 5}$, ettersom objektet antas å være en kompakt kule, med et uniformt fordelt massesenter. $y_0$ er startposisjon og $y$ er sluttposisjon.


#### Krefter på objektet:

For kulen som ruller nedover banen, vil en normalkraft og friksjonskraft virke på den. Her må det tas for seg flere komponenter. Ettersom banen har en krumning, kan den beskrives som en del av en sirkel, derav må kulen ha en sentripetalakselerasjon(Støvneng,2023,s.1). Sentripetalakselerasjonen er gitt som:

\begin{equation}
\tag{4}
a_⊥={{{2g(y_0-y)}\over {1+c}}}\cdot {y''\over [1+(y')^2]^{3\over 2}}.
\end{equation}

Når banen krummer oppover, vil $a_⊥$ være positiv, mens den vil være negativ ved krumming nedover.
Ved å ta i bruk Newtons 2. lov, har vi at normalkraften $N$ med valgt komponent $mg\cos\beta$(Støvneng,2023,s.1). Vi får da

\begin{equation}
\tag{5}
N-mg\cos\beta=a_⊥=>N={m(g\cos\beta + a_⊥)}.
\end{equation}

Ved positiv helningsvinkel vil normalkraften være positiv, og ved negtiv helningsvinkel vil normalkraften være negativ. Krumning oppover gir at $N$ og $a_⊥$ begge peker oppover. Hvis den krummer ned, vil $N$ og $a_⊥$ peke motsatt av hverandre, med $N$ pekende ned(Støvneng,2023,s.1). $\beta$ er helningsvinkelen.
For å kunne finne friksjonskraften $f$ nå, må vi ta for oss stigningstallet gitt ved:

\begin{equation}
\tag{6}
y'={dy\over dx} = tan \beta.
\end{equation}

Videre tar vi for oss den tangetielle komponenten $-mg\sin\beta$ og $f$ som virker tangentielt på banens krumning. I oppoverbakker vil da helningsvinkel og friksjon være større enn null, mens i nedoverbakker er disse mindre enn null. Retning til friksjone i oppoverbakke er mot høyre, mens nedover er mot venstre(Støvneng,2023,s.2). Ved å ta i bruk Newtons 2. lov får vi:


\begin{equation}
\tag{7}
f = {ma + mg\sin\beta = {c\over 1 + c}mg\sin\beta}
\end{equation}

hvor vi har satt inn $a={-5g\sin(\beta)\over 7}$.

Friksjonskoeffisienten:

\begin{equation}
\tag{8}
\mu={\lvert{f\over N}\lvert}.
\end{equation}

 
#### Tidsutvikling:
$t$ tid vil forsvinne i uttrykk og likninger, fordi ved bevaring av mekanisk energi utledet fra Newtons 2. lov, vil $t$-aspektet forsvinne(Støvneng,2023,s.4). Definisjonen for tidsutvikling og total rulletid kommer fra:

\begin{equation}
\tag{9}
\Delta t_n = {2 \cdot \Delta x_n \over v_{x,n-1}+v_x,_n}
\end{equation}
 

$\Delta t_n$ regner ut tid mellom intervaller fra start- til sluttposisjon. $\Delta x_n$ er det horisontalet steget, hvor det tilsvarer $\Delta x_n = x_n - x_{n-1} = 1 mm$. $v_x,_n$ tilsvarer den horisontale komponenten av hastigheten i posisjonen $x_n$(Støvneng,2023,s.4). Total rulletid vil da bli:

\begin{equation}
\tag{10}
t_n={\sum_{i=1}^{n}t_i}\;n = 1,...,1400
\end{equation}


#### Usikkerhetsanalyse:
Det vil bli avvik mellom de teoretisk og eksperimentelle verdiene. Derav tar vi i bruk middelverdi og standardfeil for å regne ut og tolke avviket.

Middelverdi, også kjent som gjennomsnitt, er størrelsen på de tilfeldige feilene når vi gjentar en måling $N$ ganger (Lilledahl & Risinggård,u.å,s.2).

\begin{equation}
\tag{11}
\bar x = {{1\over N}\sum_{i=1}^{N}x_i}.
\end{equation}



Hvor $x_i$ er listene med verdiene for totalrulletid og sluttfart for gjennomkjøringene, samt listene for total kintisk energi i sluttpunkt og tap av mekanisk energi. $N$ er antall gjennomkjøringer.
Standardfeilen er gjennomsnittet på usikkerheten. For å komme fram til standardfeilen må man først via standardavvik (Lilledahl & Risinggård,u.å,s.2). Standardavvik er oppgitt som:

\begin{equation}
\tag{12}
\Delta x = {\sqrt{{1\over N-1}\sum_{i=1}^{N}(x_i-\bar x)^2}}.
\end{equation}

Standardfeilen beregnes ved formel

\begin{equation}
\tag{13}
\Delta \bar x = {\Delta x \over \sqrt{N} }.
\end{equation}

### Diskusjon:
Når man utfører et slikt eksepriment, er det til å forvente at det vil være avvik i forhold til et nummerisk beregnet forsøk av en datamaskin. Mulige feilkilder vil være oppsett av bane, slipp av kule, tracking, kameraoppsett, koding, regnefeil og selve rullingen. I forsøket kan man skille mellom systematiske og tilfeldige feil. Systematiske feil skyldes feil i måleinstrument, uriktige avlesninger og målemetoder. Variasjon i måleresulatater som man ikke har kontroll over, er tilfeldige (Store Norske Leksikon,(2005-2007)).

#### Baneform:
Med utgangspunkt i figur 3, som er den eksperimentelle og numerisk bergende baneformen. Det er tydelig at begge representerer samme baneform. Samtidig ser vi et avvik mellom grafene festpunktene. Festepunkt 1 skal ligge på 300 mm, som er banens topphøyde. Her ligger eksperimentelle banen noe høyere. I tillegg slutter den ekseprimentelle banen tidligere enn den numeriske. Det her skyldes systematiske feil. Største mulige systematiske feilen, kan være oppsettet av banen og festing av punktene. Videre kan også kalibrering av banen i tracker ha hatt en innvirkning på hvorfor man har avvik i baneformen. 

#### Tracker:
Trackingen av de forsøkene av to personer, som hver tracket fem forsøk med egen pc. Hvilket betyr to ulike kalibrering av målestokk og koordinatsystem, og ulik nøyaktighet av plotting av punktmasse gjennom banen. Derav vil man få en forskjell i start og sluttpunkt for kulen. I og med at tracker-programmet kan identifisere punktmassene selv og tracke det, vil dette også gi en usikkerhet i nøyaktighet av verdiene. Dette grunner i at mengden av alle punktene vil variere fra forsøk til forsøk, og mellom trackingen. Tilfeldige feil kan være selve opplastning av video, avvik i videokvalitet mellom gjennomkjøringene og feil i programmet Tracker.  De systematiske feilene vil være de som ble gjennomført at de to som tracket. Baneformen og feil i tracker-programmet vil naturligvis gi følgefeil som påvirker resultat i den eksperimentelle delen av forsøket. 

#### Sluttfart og fartsgraf:
For å avgjøre rimeligheten til fartsgrafen (figur 4), sjekker man om toppunkt og bunnpunkt mellom fartsgraf og baneform stemmer overens. Toppunktene til fartsgrafen skal korrelere med bunnpunktene til baneformen, noe vi ser at det gjør. Når banen heller nedover, øker hastigheten, mens i oppoverbakke avtar den. Dette samsvarer med at farten er tilnærmert lik null i startpunket og er avtagende i sluttpunktet. Basert på figur 4, ser vi at den eksperimentelle farten ligger gjennomsnittlige lavere gjennom hele banen enn det den numeriske gjør. Systematiske feilkilder kan først og fremst være slipp av kulen. I og med at dette utføres av oss, vil slippet variere mellom forsøkene. Den kan ha fått tilført fart, blitt dratt tilbake før rulling eller en ujevn start. Dette har igjen ført til variasjone i tracking fra når kulens startpunkt er.

Energibevaring sier at energi ikke bare kan forsvinne. Den tapte mekaniske energien må ha gått over til en annen form. Dette sier oss at det for eksempel vil være luftmotstand og friksjon. Variasjonen mellom verdien for det numeriske og eksperimentelle på figur 10 er til å forvente. En datamaskin gjennomførte det numeriske forsøket, som eliminerer en del systematiske feil. #### Friksjon og normalkraft:

For å avgjøre om frisjon og normalkraft stemmer, sammenlikner vi grafene fra figur 5 med figur 3. Friksjonen har sine høyeste verdier, hvor baneformen har sine bunn- og toppunkter. I vendepunktene til banen, er friksjonen lavest. Med andre ord stemmer friksjonen overens med den eksperimentelle banen.
Ved positiv helning av banen, viser grafen til normalkraften at den er på sitt størte. Ved negativ helnig av banen, er normalkraften på sitt laveste. Dette samsvarer med teorieutledningen om normalkraft. Altså vil figur 5 være rimelig.

#### Friksjonstall:

Vi kan se på grafen (figur 6) over friksjonstallet om det er sannsynlig at vår baneform gir ren rulling. Bunnpunktene til grafen for friksjonstallet, tilsvarer topp og bunnpunkter for baneformen (figur 3). I vendepunktene til baneformen, har friksjonstallet sin største verdi. Siden vi har et friskjonstall med et maksimum på ca. $0,13$, kan vi med god margin anslå ren rulling.



### Konklusjon:
I det her forsøket har vi sett på en kule som ruller rent gjennom en bane. Med utgangspunkt i Newtons 2. lov og lov om energibevarelse, har vi gjort bergerninger, på total rulletid, sluttfart, total kinetisk energi i sluttpunkt og tap av mekanisk energi. Tilslutt sammenliknet vi de med et numerisk beregnet forsøk. Vi fikk at beregnet middelverdi og standardfeil for sluttfart ble $1,108 \pm 0,0219$, for total rulletid $2,410 \pm 0,0958$, for total kinetisk energi $0,0191 \pm 0,000754$ og for tap av mekanisk energi $0,0132 \pm 0,000754$. Derav konkluderer vi med at målte verdier er rimelige. Dersom resultatene skulle vært mer nøyaktig ville vi hadde én person tracket og vi hadde hatt flere eksperimentelle gjennomkjøringer.
